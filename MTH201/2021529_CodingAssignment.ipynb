## Question: 1
Generate 100 samples that are uniformly distributed over the interval `(0, 1)`. Plot a histogram of the samples. Also, plot an empirical CDF.

```python
import matplotlib.pyplot as plt
import numpy as np


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

count = 100

# 'count' number of samples uniformly distributed over (0, 1)
sample = np.random.uniform(0, 1, count)
n, bins, _ = axes.hist(sample, label="Samples", bins=20, rwidth=0.8)

# Normalize the samples by their total sum to get the PDF
# This PDF describes the rough probability of occurence of a value
PDF = n / count
CDF = np.cumsum(PDF)
axes.plot(bins[1:], CDF, label="CDF", linewidth=3.5)

axes.axhline(y=1, label="y = 1", color="red", linestyle="--", linewidth=1.5)

axes.grid(True)
axes.legend()
plt.show()
```

## Question: 3
Write a function that uses the uniform distribution `(0, 1)` to generate `m` samples from a Bernoulli Random Varibale with parameter `p` as input.

```python
import matplotlib.pyplot as plt
import numpy as np


def bernoulli(p: float, m: int) -> list:
    # Models 'm' repetitions of a Bernoulli Random Variable with parameter 'p'
    # np.random.random() == uniform(0, 1)
    return [int(np.random.random() < p) for _ in range(m)]


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
p = float(input("Enter the probablity of 'success' event (p): "))
samples = bernoulli(p, m)

axes.bar([0, 1], [samples.count(0), samples.count(1)], color=[str(1-p), str(p)])
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.set_xticklabels(["Failure", "Success"])
axes.set_xticks([0, 1])
plt.show()
```

## Question: 4
Write a function that generates `m` samples from a geometric Random Variable with parameter `p` as input.

```python
import matplotlib.pyplot as plt
import numpy as np


def bernoulli(p: float) -> int:
    # Models a Bernoulli Random Variable with parameter 'p'
    # np.random.random() == uniform(0, 1)
    return int(np.random.random() < p)

def geometric(p: float, m: int) -> list:
    # Models 'm' repetitions of a Geometric Random Variable with parameter 'p'
    samples = []
    for _ in range(m):
        trial = total = 0
        while trial == 0:
            trial = bernoulli(p)
            total += 1
        samples.append(total)
    return samples


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
p = float(input("Enter the probablity of 'success' event (p): "))
samples = geometric(p, m)

axes.hist(samples, np.arange(min(samples)-0.5, max(samples)+1.5, 1), rwidth=0.8)
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.set_xticks(np.arange(min(samples), max(samples)+1, 1))
axes.grid(True)
plt.show()
```

## Question: 5
Write a function that generates `m` samples from a Binomial Random Variable with parameters `n` and `p` as inputs.


```python
import matplotlib.pyplot as plt
import numpy as np


def bernoulli(p: float) -> int:
    # Models a Bernoulli Random Variable with parameter 'p'
    # np.random.random() == uniform(0, 1)
    return int(np.random.random() < p)

def binomial(n: int, p: float, m: int) -> list:
    # Models 'm' repetitions of a Binomial RV with parameters 'n' and 'p'
    return [sum(bernoulli(p) for _ in range(n)) for _ in range(m)]


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
n = int(input("Enter the number of bernoulli trials (n): "))
p = float(input("Enter the probablity of 'success' event (p): "))
samples = binomial(n, p, m)

axes.hist(samples, np.arange(min(samples)-0.5, max(samples)+1.5, 1), rwidth=0.8)
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.set_xticks(np.arange(0, max(samples)+1, 1))
axes.grid(True)
plt.show()
```

## Question: 6
Write a function that generates `m` samples from a Poisson Random Variable with parameter `alpha` as input.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp


def poisson(alpha: float, m: int) -> np.arange:
    # Models 'm' repetitions of a Poisson Random Variable with parameter 'alpha'
    return sp.stats.poisson.rvs(alpha, size=m)


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
alpha = float(input("Enter the rate of arrivals (alpha): "))
samples = poisson(alpha, m)

axes.hist(samples, np.arange(min(samples)-0.5, max(samples)+1.5, 1), rwidth=0.8)
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.set_xticks(np.arange(0, max(samples)+1, 1))
axes.grid(True)
plt.show()
```

## Question: 7
Write a function that generates `m` samples from a Gaussian Random Variable with parameters `mu` and `sigma` as input.

```python
import matplotlib.pyplot as plt
import numpy as np
import random


def gaussian(mu: float, sigma: float, m: int) -> list:
    # Models 'm' repetitions of a Gaussian RV with parameters 'mu' and 'sigma'
    return [random.gauss(mu, sigma) for _ in range(m)]


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
mu = float(input("Enter the mean (mu): "))
sigma = float(input("Enter the standard deviation (sigma): "))
samples = gaussian(mu, sigma, m)

axes.hist(samples, bins=70)
axes.axvline(x=mu, label="x = mu", color="red", linestyle="--", linewidth=1.5)
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.grid(True)
axes.legend()
plt.show()
```

## Question: 8
Write a function that generates `m` samples from an Exponential Random Variable with parameter `lamda` as input.

```python
import matplotlib.pyplot as plt
import numpy as np
import random


def exponential(lamda: float, m: int) -> list:
    # Models 'm' repetitions of an Exponential RV with parameter 'lamda'
    return [random.expovariate(lamda) for _ in range(m)]


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

m = int(input("Enter the number of samples (m): "))
lamda = float(input("Enter the rate (lamda): "))
samples = exponential(lamda, m)

axes.hist(samples, bins=70)
axes.set_xlabel("Value of the Random Variable")
axes.set_ylabel("Frequency")

axes.grid(True)
plt.show()
```

## Question: 9
Generate `m = 1, 2, ..., 1000` samples from a Bernoulli Random Variable and plot the sample mean for each `m`. Also plot the parameter `p` of the Bernoulli
Random Variable. Do you see the sample mean converge to the expected value of the Bernoulli Random Variable? For what smallest value of `m` has the sample mean more or less converged to the parameter of the Bernoulli Random Variable?

```python
import matplotlib.pyplot as plt
import random


def bernoulli(p: float) -> int:
    # Models a Bernoulli Random Variable with parameter 'p'
    # random.random() == uniform(0, 1)
    return int(random.random() < p)


fig, axes = plt.subplots(figsize=(8, 5))
axes.axhline(y=0, color="black", linewidth=3)
axes.axvline(x=0, color="black", linewidth=3)
axes.plot(0, -0.2, color="black")

p = float(input("Enter the probability of 'success' event of Bernoulli (p): "))
X, Y, means = [0, 1], [0, 0], []

for m in range(1, 1001):
    trial = bernoulli(p)
    Y[trial] += 1
    means.append(Y[1] / m)
    color = str(p) if trial else str(1-p)
    axes.plot(m, trial, color=color, marker=".")

axes.axhline(y=p, label="y = E[X] = p", color="red", linestyle="--", linewidth=2)
axes.plot(m, 0, label="Failures", color=str(1-p), marker=".")
axes.plot(m, 1, label="Successes", color=str(p), marker=".")
axes.plot(range(1, 1001), means, linewidth=2)

axes.set_xlabel("Number of Trials")
axes.set_ylabel("Mean")

axes.grid(True)
axes.legend()
plt.show()
```

### Answer to Question: 9
On an average, the mean after a large number of Bernoulli Trials converges its Expected Value, `E[X] = p`. <br>
After the number of trials cross `~250`, the mean gets more or less converged to the parameter of the Random Variable.
